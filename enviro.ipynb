```python
("i'm learning")
temprature = 23
(temprature)
day1= 42
day2= 55
day3= 61
Average = (day1 + day2 + day3) / 3
(Average)

lake_ph =[6.5, 6.8, 7.0, 6.9, 6.7]
print(lake_ph)
average_ph= sum(lake_ph) / len(lake_ph)
print(average_ph)
rounded_ph = round(average_ph, 2)
print(average_ph)
!pip install matplotlib
import matplotlib.pyplot as plt
lake_ph = [6.5, 6.8, 7.0, 6.9, 6.7]
days = [1, 2, 3, 4, 5]
plt.plot(days, lake_ph, color = 'red')
plt.title( 'lake pH over 5 days')
plt.xlabel('day')
plt.ylabel('ph level')
plt.grid(True)
import matplotlib.pyplot as plt

lake_ph = [6.5, 6.8, 7.0, 6.9, 6.7]
days = [1, 2, 3, 4, 5]

plt.plot(days, lake_ph, color='green', linestyle='--', marker='s', markerfacecolor='blue', markersize=8)
plt.title('Lake pH Over 5 Days')
plt.xlabel('Day')
plt.ylabel('pH Level')
plt.grid(True)
plt.legend()
plt.show()
!pip install pandas
import pandas as pd
url = "https://raw.githubusercontent.com/selva86/datasets/master/AirPassengers.csv"
df = pd.read_csv(url)
df.head()
import matplotlib.pyplot as plt
df.columns = ['Month', 'Pollution']
plt.figure(figsize=(10, 5))
plt.plot(df['Month'], df['Pollution'])
plt.title('pm2.5 Pollution Over Time')
plt.xlabel('Month')
plt.ylabel('pm2.5 concentration')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
df['Month'] = pd.to_datetime(df['Month'])
df_1950 = df[df['Month'].dt.year >= 1950]
plt.figure(figsize=(8, 4))
plt.plot(df_1950['Month'], df_1950['Pollution'])
plt.title('Pollution in 1950')
plt.xlabel('Month')
plt.ylabel('pm2.5 concentration')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
max_row = df.loc[df['Pollution'].idxmax()]
print("highest_pollution:" , max_row['Month'].date(), "with value", max_row['Pollution'])
min_row = df.loc[df['Pollution'].idxmin()]
print("lowest_pollution:" , min_row['Month'].date(), "with value", min_row['Pollution'])
import pandas as pd
import numpy as np

data = {
    'Date': [
        '2025-08-01',
        '2025-08-02',
        '2025-08-03',
        '2025-08-04',
        '2025-08-05',
        '2025-08-06'
    ],
    'Temperature': [
        72,
        None,
        75,
        'error',
        78,
        77
    ],
    'Humidity': [
        55,
        58,
        None,
        59,
        61,
        'missing'
    ]
}


df_dirty = pd.DataFrame(data)


df_dirty.replace(['error', 'missing'], np.nan, inplace=True)

df_dirty['Temperature'] = pd.to_numeric(df_dirty['Temperature'], errors='coerce')
df_dirty['Humidity'] = pd.to_numeric(df_dirty['Humidity'], errors='coerce')

df_cleaned = df_dirty.dropna()

df_cleaned

avg_temp = df_cleaned['Temperature'].mean()
avg_humidity = df_cleaned['Humidity'].mean()

max_temp = df_cleaned['Temperature'].max()
min_temp = df_cleaned['Temperature'].min()

max_humidity = df_cleaned['Humidity'].max()
min_humidity = df_cleaned['Humidity'].min()

print("Average Temperature:", round(avg_temp, 2))
print("Average Humidity:", round(avg_humidity, 2))
print("Highest Temp:", max_temp)
print("Lowest Temp:", min_temp)
print("Highest Humidity:", max_humidity)
print("Lowest Humidity:", min_humidity)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(df_cleaned['Date'], df_cleaned['Temperature'], marker='o', label='Temperature (°F)', color='orange')
plt.plot(df_cleaned['Date'], df_cleaned['Humidity'], marker='s', label='Humidity (%)', color='blue')

plt.title('Environmental Conditions Over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
